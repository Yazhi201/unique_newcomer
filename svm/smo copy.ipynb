```python
def loadDataSet(filename):
    fr = open(filename)
    data = []
    label = []
    for line in fr.readlines():
        lineAttr = line.strip().split('\t')
        data.append([float(x) for x in lineAttr[:-1]])
        label.append(float(lineAttr[-1]))
    return data,label
```

```python
def selectJrand(i, m):
    j = i
    while j == i:
        j = int(random.uniform(0, m))
    return j
```

```python
def clipAlpha(a_j, H, L):
    if a_j > H:
        a_j = H
    if L > a_j:
        a_j = L
    return a_j
```

```python
def Get_LH(li, lj, C, ai, aj):
    if li != lj:
        L = max(0, aj - ai)
        H = min(C, C+aj-ai)
    else:
        L = max(0, ai+aj - C)
        H = min(C, aj+ai)
    return L, H
```

```python
def _K(data, i, j):
    return (data[i, :]*data[j, :].T) ** 2
```

```python
def smo(T_data, T_label, C, toler, maxIter):
    data = mat(T_data)
    label = mat(T_label).transpose()
    # exit(0)
    b = 0.0
    iter = 0
    m, n = shape(data)
    alpha = mat(zeros((m, 1)))
    while iter < maxIter:
        changenum = 0
        for i in range(m):
            # print(multiply(alpha, label).T.shape)
            # print((data * data[i, :].T).shape)
            fxi = float(multiply(alpha, label).T * (data * data[i, :].T)) + b
            Ei = fxi - float(label[i])
            if label[i] * Ei < -toler and alpha[i] < C or label[i] * Ei > toler and alpha[i] > 0:
                j = selectJrand(i, m)
                fxj = float(multiply(alpha, label).T *
                            (data * data[j, :].T)) + b
                Ej = fxj - float(label[j])
                alphaIOld = alpha[i].copy()
                alphaJOld = alpha[j].copy()
                L, H = Get_LH(label[i], label[j], C, alpha[i], alpha[j])
                if L == H:
                    # print ("L==H")
                    continue
                eta = 2.0 * _K(data, i, j) - _K(data, i, i) - \
                    _K(data, j, j)
                if eta >= 0:
                    # print ("eta >= 0")
                    continue
                alpha[j] -= label[j]*(Ei-Ej)/eta
                alpha[j] = clipAlpha(alpha[j], H, L)
                if abs(alpha[j] - alphaJOld) < 0.00001:
                    # print ("j not move enough")
                    continue
                alpha[i] += label[j]*label[i]*(alphaJOld - alpha[j])
                b1 = b - Ei - label[i]*(alpha[i] - alphaIOld)*_K(data, i, i) \
                    - label[j]*(alpha[j]-alphaJOld)*_K(data, i, j)
                b2 = b - Ej - label[i]*(alpha[i] - alphaIOld)*_K(data, i, j) \
                    - label[j]*(alpha[j]-alphaJOld)*_K(data, j, j)
                if alpha[i] > 0 and alpha[i] < C:
                    b = b1
                elif alpha[j] > 0 and alpha[j] < C:
                    b = b2
                else:
                    b = (b1+b2)/2.0
                changenum += 1
        if changenum == 0:
            iter += 1
        else:
            iter = 0
        print("iteration number: %d" % iter)
    return b, alpha
```

```python
data, label = loadDataSet('test1.txt')
# print (data)
# print (label)
for i in range(size(label)):
    if label[i] == 0:
        label[i] = -1

b, alpha = smo(data, label, 0.6, 0.001, 40)
```
