```python
import numpy as np
import pandas as pd
```

```python
train_df = pd.read_csv("3.csv")
train_df = train_df.drop(['PassengerId'], axis=1)
print(train_df.head)
idx = np.random.permutation(train_df.shape[0])
# print(idx)
train_df = train_df.iloc[idx]
test_df, train_df = train_df[0 : 200], train_df[200 : ]
```

```python
# 划分好了东西，现在开始随机梯度下降
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))


def get_acc(y, y_pred): 
    ans = 0
    for j in range(y.shape[0]):
        # print(y_pred[j], y[j][0])
        ans += (y_pred[j] == y[j][0])
    return ans / y.shape[0]
```

### LR

```python
class LogisticRegression():

    def __init__(self, learning_rate=0.00001, n_iterations=30000, C = 0.001):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.C = C

    def initialize_weights(self, n_features):
        # 初始化参数
        # 参数范围[-1/sqrt(N), 1/sqrt(N)]
        limit = np.sqrt(1 / n_features)
        w = np.random.uniform(-limit, limit, (n_features, 1))
        b = 0
        self.w = np.insert(w, 0, b, axis=0)

    def fit(self, X, y):
        m_samples, n_features = X.shape
        self.initialize_weights(n_features)
        # 为X增加一列特征x1，x1 = 0
        X = np.insert(X, 0, 1, axis=1)
        y = np.reshape(y, (m_samples, 1))

        # 梯度训练n_iterations轮
        for i in range(self.n_iterations):
            h_x = X.dot(self.w)
            y_pred = sigmoid(h_x)
            if i % 100 == 0:
                ans = 0
                # print((y_pred.astype(int)).shape)
                for k in range(y.shape[0]):
                    flag = 0
                    if y_pred[k] >= 0.5:
                        flag=1
                    ans += flag == y[k]
                print("test: {} acc: ".format(i),
                    ans / y.shape[0])
                loss = (-1.0 / y.shape[0]) * np.sum(y * np.log(y_pred + 1e-20) + (
                    1 - y) * np.log(1 - y_pred + 1e-20)) - np.matmul(self.w.T, self.w)[0][0] * self.C
                print("Loss: {}".format(i), loss)
            print(self.w.shape)
            w_grad = X.T.dot(y_pred - y) + self.w * self.C
            self.w = self.w - self.learning_rate * w_grad


    def predict(self, X):
        X = np.insert(X, 0, 1, axis=1)
        print(X.shape)
        h_x = X.dot(self.w)
        y_pred = np.round(sigmoid(h_x))
        print(y_pred.shape)
        return y_pred.astype(int)
```

```python
LR = LogisticRegression()
train_np = train_df.values
test_np = test_df.values
LR.fit(train_np[:,1:], train_np[:,0])


```

```python
y_pred = LR.predict(test_np[:, 1:])
print(type(get_acc(test_np, y_pred)))
print("acc: ", get_acc(test_np, y_pred))
```

## SGD

```python
class LogisticRegression_SGD():

    def __init__(self, learning_rate=0.00001, n_iterations=300000, C=0.001):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.C = C

    def initialize_weights(self, n_features):
        # 初始化参数
        # 参数范围[-1/sqrt(N), 1/sqrt(N)]
        limit = np.sqrt(1 / n_features)
        w = np.random.uniform(-limit, limit, (n_features, 1))
        b = 0
        self.w = np.insert(w, 0, b, axis=0)

    def fit(self, X, y):
        m_samples, n_features = X.shape
        self.initialize_weights(n_features)
        # 为X增加一列特征x1，x1 = 0
        X = np.insert(X, 0, 1, axis=1)
        y = np.reshape(y, (m_samples, 1))

        # 梯度训练n_iterations轮
        for i in range(self.n_iterations):
            # SGD随机挑选一些东西来用

            my_per = np.random.permutation(X.shape[0])
            np.random.shuffle(my_per)
            my_per = my_per[:100]

            XX = X[my_per]
            yy = y[my_per]
            # print(yy.shape)
            # print(XX.shape)
            h_x = XX.dot(self.w)
            
            y_pred = sigmoid(h_x)
            if i % 100 == 0:
                ans = 0
                # print((y_pred.astype(int)).shape)
                for k in range(yy.shape[0]):
                    flag = 0
                    if y_pred[k] >= 0.5:
                        flag = 1
                    ans += flag == y[k]
                print("test: {} acc: ".format(i),
                      ans / yy.shape[0])
                loss = (-1.0 / yy.shape[0]) * np.sum(yy * np.log(y_pred + 1e-20) + (
                    1 - yy) * np.log(1 - y_pred + 1e-20)) - np.matmul(self.w.T, self.w)[0][0] * self.C
                print("Loss: {}".format(i), loss)

            w_grad = XX.T.dot(y_pred - yy) + self.w * self.C
            self.w = self.w - self.learning_rate * w_grad

    def predict(self, X):
        X = np.insert(X, 0, 1, axis=1)
        h_x = X.dot(self.w)
        y_pred = np.round(sigmoid(h_x))
        return y_pred.astype(int)
```

## Adam

```python
class LogisticRegression_Adam():

    def __init__(self, learning_rate=0.00001, n_iterations=30000, C=0.001):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.C = C
        self.alpha = 0.8
        self.belta1 = 0.999
        self.belta2 = 0.999
        self.ep = 0.5
        self.v = np.zeros(1)
        self.m = np.zeros(1)


    def initialize_weights(self, n_features):
        # 初始化参数
        # 参数范围[-1/sqrt(N), 1/sqrt(N)]
        limit = np.sqrt(1 / n_features)
        w = np.random.uniform(-limit, limit, (n_features, 1))
        b = 0
        self.w = np.insert(w, 0, b, axis=0)

    def updata (self, w_grad, t):
        m1 = self.m
        v1 = self.v
       
        self.m = m1 * self.belta1 + w_grad * (1 - self.belta1)
        self.v = v1 * self.belta2 + (w_grad * w_grad) * (1 - self.belta2)
        # print(self.m, self.v)
        # print(t )
        t += 1
        # print("sda",(1 - (self.belta1 ** t)))
        m2 = self.m / (1 - (self.belta1 ** t))
        # print(m2)
        v2 = self.v / (1 - (self.belta2 ** t))
        # print(v2)
        lp = self.alpha * m2 / (np.sqrt(v2) + self.ep)
        # print((w_grad * (1 - self.belta1)).shape)
        # print((m1 * self.belta1 + w_grad * (1 - self.belta1)).shape)
        # print("lp,", lp.shape)
        # print(lp)
        return lp




    def fit(self, X, y):
        m_samples, n_features = X.shape
        self.initialize_weights(n_features)
        # 为X增加一列特征x1，x1 = 0
        X = np.insert(X, 0, 1, axis=1)
        y = np.reshape(y, (m_samples, 1))

        # 梯度训练n_iterations轮

        self.m = np.zeros((X.shape[1],1))
        self.v = np.zeros((X.shape[1], 1))
        # print(self.m.shape)
        for i in range(self.n_iterations):
        # for i in range(1):
            h_x = X.dot(self.w)
            y_pred = sigmoid(h_x)
            if i % 100 == 0:
                ans = 0
                # print((y_pred.astype(int)).shape)
                for k in range(y.shape[0]):
                    flag = 0
                    if y_pred[k] >= 0.5:
                        flag = 1
                    ans += flag == y[k]
                print("test: {} acc: ".format(i),
                      ans / y.shape[0])
                loss = (-1.0 / y.shape[0]) * np.sum(y * np.log(y_pred + 1e-20) + (
                    1 - y) * np.log(1 - y_pred + 1e-20)) - np.matmul(self.w.T, self.w)[0][0] * self.C
                print("Loss: {}".format(i), loss)

            w_grad = X.T.dot(y_pred - y) + 2 * self.w * self.C   
            # print(w_grad.shape)
            # print(self.w.shape)
            # exit(0)
            # print(self.updata(w_grad, i))
            # print(w_grad)
            w_grad = self.updata(w_grad, i)
            self.w = self.w - w_grad
            # print(self.w)
            # print(w_grad)
            # print(self.w.shape)

            
            # self.w = self.w - self.learning_rate * w_grad

    def predict(self, X):
        X = np.insert(X, 0, 1, axis=1)
        h_x = X.dot(self.w)
        y_pred = np.round(sigmoid(h_x))
        return y_pred.astype(int)
```
